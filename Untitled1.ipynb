```python
# Cell 1: Import Library
import os
import glob
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
from sklearn.model_selection import train_test_split
```

## Pendekatan Pemecahan Masalah

Dataset berisi:
- Image
- YOLO bounding box annotation
- **Belum ada label warna**, sehingga kita harus menentukan label sendiri.

### Langkah-langkah utama:
1. Membaca gambar dan bounding box.
2. Labeling manual menggunakan roboflow
3. Mengklasifikasikan menjadi:
   - `light_blue`
   - `dark_blue`
   - `others`
4. Melatih model YOLOv8.
5. Mengevaluasi dan menganalisis hasil.

### Kenapa YOLOv8n?
- Sangat ringan → cocok untuk edge device
- Akurasi cukup baik untuk data sederhana
- Cepat inference <10ms per frame


```python
# =========================================================
# Cell 2: Definisikan path dataset dan tampilkan 3 contoh gambar
IMAGE_PATH = "/content/data/images"
LABEL_PATH = "/content/data/labels"

# Ambil semua file jpg
all_images = glob.glob(os.path.join(IMAGE_PATH, "*.jpg"))

# Tampilkan 3 contoh gambar
plt.figure(figsize=(15,5))
for i, img_file in enumerate(all_images[:3]):
    img = cv2.imread(img_file)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.subplot(1, 3, i+1)
    plt.imshow(img_rgb)
    plt.axis("off")
    plt.title(f"Contoh {i+1}")
plt.show()
```

```python
# Cell 3: Fungsi konversi YOLO ke pixel
def yolo_to_pixel(yolo_bbox, img_w, img_h):
    cls, xc, yc, bw, bh = yolo_bbox
    x1 = int((xc - bw/2) * img_w)
    x2 = int((xc + bw/2) * img_w)
    y1 = int((yc - bh/2) * img_h)
    y2 = int((yc + bh/2) * img_h)
    return int(cls), x1, y1, x2, y2
```

```python
# Cell 4: Visualisasi beberapa bounding box
sample_images = glob.glob(os.path.join(IMAGE_PATH, "*.jpg"))[:5]

for img_file in sample_images:
    img = cv2.imread(img_file)
    h, w = img.shape[:2]

    label_file = os.path.join(LABEL_PATH, os.path.basename(img_file).replace(".jpg",".txt"))
    if not os.path.exists(label_file):
        continue

    with open(label_file, "r") as f:
        lines = f.readlines()

    for line in lines:
        cls, xc, yc, bw, bh = map(float, line.strip().split())
        x1 = int((xc - bw/2) * w)
        x2 = int((xc + bw/2) * w)
        y1 = int((yc - bh/2) * h)
        y2 = int((yc + bh/2) * h)

        # warna untuk visualisasi
        color = ["dark_blue","light_blue","other"][int(cls)]
        color_bgr = (255,255,0) if color=="light_blue" else (255,0,0) if color=="dark_blue" else (128,128,128)
        cv2.rectangle(img, (x1,y1), (x2,y2), color_bgr, 2)
        cv2.putText(img, color, (x1,y1-5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color_bgr, 1)

    plt.figure(figsize=(6,6))
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()
```

```python
# Cell 5: Hitung jumlah label per kelas
counts = {0:0, 1:0, 2:0}
for label_file in glob.glob(os.path.join(LABEL_PATH, "*.txt")):
    with open(label_file,"r") as f:
        for line in f:
            cls = int(line.strip().split()[0])
            counts[cls] += 1

plt.bar(["light_blue","dark_blue","other"], counts.values(), color=['skyblue','blue','gray'])
plt.title("Jumlah Label per Kelas")
plt.ylabel("Jumlah")
plt.show()
```

```python
# Cell 6: Split dataset
all_images = glob.glob(os.path.join(IMAGE_PATH,"*.jpg"))
train_imgs, val_imgs = train_test_split(all_images, test_size=0.2, random_state=42)

with open("/content/train.txt","w") as f:
    for p in train_imgs:
        f.write(p+"\n")
with open("/content/val.txt","w") as f:
    for p in val_imgs:
        f.write(p+"\n")

print(f"Jumlah train: {len(train_imgs)}, val: {len(val_imgs)}")
```

```python
import yaml

# Create data.yaml file for YOLOv8 training
data_config = {
    "train": "/content/train.txt",
    "val": "/content/val.txt",
    "nc": 3,
    "names": ["dark_blue", "light_blue", "other"]
}

with open("/content/data.yaml", "w") as f:
    yaml.dump(data_config, f)

print("data.yaml created successfully:")
with open("/content/data.yaml", "r") as f:
    print(f.read())
```

```python
# Cell 7: Training YOLOv8 nano
model = YOLO("yolov8n.pt")  # pretrained YOLOv8 nano
model.train(
    data="/content/data.yaml",  # Pass the path to the YAML file
    epochs=20,
    imgsz=640,
    batch=16,
    device="cpu"  # pakai CPU karena GPU tidak tersedia
)
```

```python
# Cell 8: Evaluasi Model (mAP & Prediksi Visual)
# Load model yang sudah di-training
model = YOLO("/content/runs/detect/train2/weights/best.pt")  # sesuaikan path jika berbeda

# Hitung metrics pada validation set
metrics = model.val(data="/content/data.yaml")
```

```python
# =========================================================
# Cell 9: Visualisasi Prediksi Model dengan Bounding Box

import cv2
import matplotlib.pyplot as plt
import numpy as np

sample_val_images = val_imgs[:5]  # ambil 5 gambar dari validation

# Mapping warna dan label
label_colors = {
    0: (255, 255, 0),   # light_blue - kuning
    1: (255, 0, 0),     # dark_blue - biru
    2: (128, 128, 128)  # other - abu-abu
}

label_names = {
    0: "light_blue",
    1: "dark_blue",
    2: "other"
}

for img_file in sample_val_images:
    # Load gambar
    img = cv2.imread(img_file)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Prediksi model
    results = model.predict(img_file, imgsz=640, conf=0.25)  # pastikan conf threshold cukup rendah

    # YOLOv8 terkadang mengembalikan list, ambil hasil pertama
    r = results[0]

    # Pastikan ada boxes
    if hasattr(r, "boxes") and len(r.boxes) > 0:
        boxes = r.boxes.xyxy.cpu().numpy()  # [x1, y1, x2, y2]
        confs = r.boxes.conf.cpu().numpy()
        cls_ids = r.boxes.cls.cpu().numpy().astype(int)

        for box, conf, cls_id in zip(boxes, confs, cls_ids):
            x1, y1, x2, y2 = map(int, box)
            color_bgr = label_colors[cls_id]
            label_text = f"{label_names[cls_id]} {conf:.2f}"

            # Draw bounding box
            cv2.rectangle(img_rgb, (x1, y1), (x2, y2), color_bgr, 2)

            # Draw label background
            (text_w, text_h), _ = cv2.getTextSize(label_text, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 1)
            cv2.rectangle(img_rgb, (x1, y1 - text_h - 4), (x1 + text_w, y1), color_bgr, -1)

            # Draw text
            cv2.putText(img_rgb, label_text, (x1, y1 - 2),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 1, cv2.LINE_AA)

    # Tampilkan gambar
    plt.figure(figsize=(6,6))
    plt.imshow(img_rgb)
    plt.axis("off")
    plt.title(f"Prediksi Model: {img_file.split('/')[-1]}")
    plt.show()
```

```python
# Cell 10: Hitung FPS Inference
import time

test_img = cv2.imread(sample_val_images[0])
start = time.time()
for _ in range(50):  # lakukan 50 kali untuk rata-rata
    _ = model.predict(test_img, imgsz=640)
end = time.time()
fps = 50 / (end - start)
print(f"Inference FPS: {fps:.2f} frame per second")
print(f"Estimasi inference time per frame: {1000/fps:.2f} ms")
```
